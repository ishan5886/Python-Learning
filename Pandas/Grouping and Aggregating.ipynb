```python
pythonresp = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
python_df = pd.concat([totalresp, pythonresp], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country' : 'TotalResp', 'LanguageWorkedWith': 'PythonUsers'}, inplace=True)
```

```python
python_df
```

```python
python_df['PythonPercentage'] = (python_df['PythonUsers'] / python_df['TotalResp']) * 100
```

```python
python_df
```

```python
python_df.sort_values(by='PythonPercentage', inplace=True)
```
